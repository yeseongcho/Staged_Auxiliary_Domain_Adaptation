#### (2) Average of Sharpness in Clean Cityscapes

```python
def compute_texture_sharpness(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Calculate gradient by using Sobel filter
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
    
    # Get gradients (sharpness)
    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    
    # Average of sharpness
    sharpness_score = np.mean(gradient_magnitude)
    
    return sharpness_score
```

#### (3) Decision Threshold

```python
import cv2
import numpy as np

def z_score_normalization(data):
    mean = np.mean(data)    
    std = np.std(data)      
    z_score_normalized_data = (data - mean) / std    
    return z_score_normalized_data

def check_threshold(Y_score, sobel_score) :
    threshold_Y = np.array([78.79, 99.86, 114.74, 136.31])
    threshold_sobel = np.array([75.28, 79.52, 82.37, 86.12]) ## After applied -> 100 - sharpness_threshold

    mean_Y = np.mean(threshold_Y)
    std_Y = np.std(threshold_Y)

    mean_sobel = np.mean(threshold_sobel)
    std_sobel = np.std(threshold_sobel)

    ## Normalization
    z_threshold_Y= z_score_normalization(threshold_Y)
    z_threshold_sobel =  z_score_normalization(threshold_sobel)

    ## Average of Luminance and Sharpness to define final threshold
    final_threshold = np.mean([z_threshold_Y,z_threshold_sobel],axis=0)

    ## Normalization of Input 
    input_Y_z_score = (Y_score-mean_Y)/std_Y
    input_sobel_z_score = (sobel_score-mean_sobel)/std_sobel

    final_score = (input_Y_z_score+input_sobel_z_score)/2

    clear_light_middle_threshold = (final_threshold[0]+final_threshold[1])/2

    medium_dense_middle_threshold = (final_threshold[2]+final_threshold[3])/2

    ### Final Decision 
    if final_score < final_threshold[0] :
        answer = "Clear"
    elif final_score >= final_threshold[0] and final_score < clear_light_middle_threshold :
        answer = "Light Haze"
    elif final_score >= clear_light_middle_threshold and final_score < final_threshold[1] :
        answer = "Light Haze"
    elif final_score >= final_threshold[1] and final_score < 0 :
        answer = "Light Haze"
    elif final_score >= 0 and final_score < final_threshold[2] :
        answer = "Medium Haze"
    elif final_score >= final_threshold[2] and final_score < medium_dense_middle_threshold :
        answer = "Medium Haze"
    elif final_score >= medium_dense_middle_threshold and final_score < final_threshold[3] :
        answer = "Dense Haze"
    elif final_score >= final_threshold[3] :
        answer = "Dense Haze"
        
    return [final_threshold[0],clear_light_middle_threshold,final_threshold[1],0,final_threshold[2],medium_dense_middle_threshold,final_threshold[3]] , final_score, answer
```

```python
def calculate_Y_mean(rgb_image) : 

    ycbcr_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2YCrCb)

    Y, Cb, Cr = cv2.split(ycbcr_image)

    y_mean = np.mean(Y)

    return y_mean

def calculate_sobel_mean(rgb_image) : 
    gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)

    gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    sobel_mean = 100-np.mean(gradient_magnitude)

    return sobel_mean
```
